Importing Libraries 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random as rd
```

Generating random data for the Social Media Data

```python
categories = ['Food', 'Travel', 'Fashion', 'Fitness', 'Music', 'Culture', 'Family', 'Health', 'Politics', 'Technology']
data = pd.DataFrame(columns=['Category', 'Views', 'Likes', 'Date', 'Comments', 'Dislikes' ])
date = pd.date_range(start='1/1/2019', end='1/1/2020', periods = 500)
categories = rd.choices(categories, k=500)
likes = np.random.randint(0, 100000, 500)
dislikes = np.random.randint(0, 100000, 500)
views = np.random.randint(0, 1000000, 500)
comments = np.random.randint(0, 100000, 500)
```

Loading the data into a Pandas DataFrame and Exploring the Data

```python
df = pd.DataFrame({'Category': categories, 'Views': views, 'Likes': likes, 'Date': date, 'Comments': comments, 'Dislikes': dislikes })
print(df)
```

Cleaning the Data

```python
v = df['Views'].mean()
l = df['Likes'].mean()
dl= df['Dislikes'].mean()
c = df['Comments'].mean()
df['Views'].fillna(v, inplace=True)
df['Likes'].fillna(l, inplace=True)
df['Dislikes'].fillna(dl, inplace=True)
df['Comments'].fillna(c, inplace=True)
print(df)
```

```python
pd.to_datetime(df['Date'])
print(df)
```

Visualizing and Analyzing the data

```python
sns.histplot(df['Likes'])
plt.xlabel('Likes')
plt.ylabel('Frequency')
plt.title('Histogram of Likes')
plt.show()
```

Creating a Boxplot

```python
sns.boxplot(x='Category', y='Likes', data=df)
plt.xticks(rotation=45)
plt.xlabel('Category')
plt.ylabel('Likes')
plt.title('Boxplot of Likes by Category')
plt.show()
mean_likes_by_category = df.groupby('Category')['Likes'].mean()
print(mean_likes_by_category)
```
